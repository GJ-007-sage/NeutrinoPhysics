```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')  # Professional style, fallback: 'ggplot' or 'fivethirtyeight'

# Parameters
delta_m2 = 2.5e-3  # eV^2, SK best-fit (Page 43)
sin2_2theta = 0.9
E_nu = 1  # GeV
L = np.linspace(0, 10000, 1000)  # km

# Oscillation probabilities
P_mu_tau = sin2_2theta * np.sin(1.27 * delta_m2 * L / E_nu)**2
P_mu_mu = 1 - P_mu_tau

# Calculate oscillation length and maximum
L_osc = 2.48 * E_nu / delta_m2
L_max = np.pi * E_nu / (1.27 * delta_m2)
print(f"Oscillation Length (Δm² = 2.5e-3 eV²): {L_osc:.0f} km")
print(f"First Maximum (Δm² = 2.5e-3 eV²): {L_max:.0f} km")

# Plot
plt.figure(figsize=(12, 7))
plt.plot(L, P_mu_tau, label=r'$P(\nu_\mu \to \nu_\tau)$, $\Delta m^2 = 2.5 \times 10^{-3}$ eV$^2$', color='blue', linewidth=2)
plt.plot(L, P_mu_mu, label=r'$P(\nu_\mu \to \nu_\mu)$, $\Delta m^2 = 2.5 \times 10^{-3}$ eV$^2$', color='red', linewidth=2)
plt.xlabel('Pathlength $L$ (km)', fontsize=12)
plt.ylabel('Oscillation Probability', fontsize=12)
plt.title('Two-Flavor Neutrino Oscillation ($E_\\nu = 1$ GeV)', fontsize=14)
plt.grid(True)
plt.legend(fontsize=10, loc='upper right', facecolor='white', edgecolor='black', framealpha=1)
plt.savefig('oscillation_probability_clean.png', dpi=300)
plt.show()
```
