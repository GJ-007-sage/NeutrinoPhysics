```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')  # Professional style

# Parameters
sin2_theta_values = [0.1, 0.05, 0.01]  # Vacuum mixing angles
colors = ['blue', 'red', 'green']  # Colors for different angles

# Xi array (dimensionless)
xi = np.linspace(-5, 5, 1000)

# Effective mixing angle in matter
def sin2_2theta_m(xi, sin2_theta):
    theta = np.arcsin(np.sqrt(sin2_theta))
    sin2_2theta = np.sin(2 * theta)**2
    cos_2theta = np.cos(2 * theta)
    denominator = (xi - cos_2theta)**2 + sin2_2theta
    return sin2_2theta / denominator

# Create plot
plt.figure(figsize=(12, 7))
for sin2_theta, color in zip(sin2_theta_values, colors):
    sin2_2theta_m_values = sin2_2theta_m(xi, sin2_theta)
    theta = np.arcsin(np.sqrt(sin2_theta))
    sin2_2theta = np.sin(2 * theta)**2
    cos_2theta = np.cos(2 * theta)
    label = r'$\sin^2 \theta = ' + f'{sin2_theta:.3f}$'
    plt.plot(xi, sin2_2theta_m_values, label=label, color=color, linewidth=2)
    plt.axhline(y=sin2_2theta, color=color, linestyle='--', alpha=0.5)
    plt.axvline(x=cos_2theta, color=color, linestyle=':', alpha=0.5)
    # Annotate resonance point
    max_y = sin2_2theta_m(cos_2theta, sin2_theta)
    plt.annotate(r'Resonance $\xi \approx ' + f'{cos_2theta:.2f}$', 
                 xy=(cos_2theta, max_y), xytext=(cos_2theta + 1*cos_2theta , max_y - 0.5*cos_2theta),
                 arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=5),
                 fontsize=8, color=color)

plt.xlabel(r'$\xi = 2 V E / \Delta m^2$', fontsize=12)
plt.ylabel(r'Effective Mixing Parameter $\sin^2 2\theta_m$', fontsize=12)
plt.title(r'MSW Resonance: Effective Mixing Angle for Two-Flavor Mixing', fontsize=14)
plt.grid(True)
plt.legend(fontsize=10, loc='upper right', facecolor='white', edgecolor='black', framealpha=1)
plt.ylim(0, 1.1)
plt.xlim(-4, 4)
# Optional: Uncomment for log scale (requires positive xi)
# plt.xscale('log')
# plt.xlim(1e-3, 5)

plt.show()
```
